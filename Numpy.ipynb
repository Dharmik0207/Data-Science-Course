### Vectorization & Broadcasting

```python
#Vectorization
arr = np.array([23, 4, 34, 45])
```

```python
result = arr ** 2
```

```python
print(result)
```

```python
#Broadcasting
result = arr + 10
```

```python
print(result)
```

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([1, 2, 3])
```

```python
result = arr1 + arr2
```

```python
print(result)
```
