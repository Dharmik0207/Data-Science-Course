### Binomial Distribution :

```python
from scipy.stats import binom
```

```python
# Probability of exactly 6 heads in 10 tosses (p = 0.5)
prob = binom.pmf(k=6, n=10, p=0.5)
print(prob)
```

### Normal Distribution :

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate values
x = np.linspace(-4, 4, 1000) # linspace(start, end, number_of_points)
mean = 0 # Mean (μ) is the center of the distribution. For standard normal distribution μ = 0.
std_dev = 1 # Standard deviation (σ) controls spread of the curve. (σ = 1)
```

```python
#  Get the probability density
y = norm.pdf(x, loc=mean, scale=std_dev)        # Parameter	Meaning
                                                # x	values where probability is calculated
                                                # loc=mean	center of distribution
                                                # scale=std_dev	spread of distribution
```

```python
# plot
plt.plot(x, y)
plt.title("Standard Normal Distribution (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True) # Adds grid lines to make the graph easier to read.
plt.show() # Without this line, the graph may not appear.
```

```python
x = np.linspace(-4, 8, 1000)
mean = 2
std = 1.5

y = norm.pdf(x, loc=mean, scale=std)

plt.plot(x, y)
plt.title("Standard Normal Distribution (μ={mean}, σ={std})")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```
